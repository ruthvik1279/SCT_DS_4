```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

csv_path = "bank-full.csv"
df = pd.read_csv(csv_path, sep=";")

sns.set(style="whitegrid")

plt.figure(figsize=(10,4))
sns.countplot(x="job", data=df, order=df['job'].value_counts().index, hue="job", palette="Blues", legend=False)
plt.xticks(rotation=45)
plt.title("Distribution of Customers by Job")
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(x="marital", data=df, hue="marital", palette="muted", legend=False)
plt.title("Distribution by Marital Status")
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(x="education", data=df, hue="education", order=df['education'].value_counts().index, palette="Set2", legend=False)
plt.title("Distribution by Education")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(x="housing", data=df, hue="y", palette="coolwarm")
plt.title("Housing Loan vs Subscription")
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(x="loan", data=df, hue="y", palette="pastel")
plt.title("Personal Loan vs Subscription")
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(x="y", data=df, hue="y", palette="Reds", legend=False)
plt.title("Subscription Outcome (Target Variable)")
plt.show()

numeric_cols = ["age","balance","duration","campaign","previous"]
plt.figure(figsize=(8,6))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap (Numeric Features)")
plt.show()
```
